# Paso 2: quitar valores atípicos

```python
for i in range(len(DB)):
    if DB['provedores_nacionales'][i]>100:
        DB.loc[i,'provedores_nacionales'] = np.nan
        
    if DB['ventas_porcentaje'][i]>100:
        DB.loc[i,'ventas_porcentaje'] = np.nan
        
    if DB['perdidas_porcentaje'][i]>100:
        DB.loc[i,'perdidas_porcentaje'] = np.nan
        
    if DB['capacidad'][i]>100:
        DB.loc[i,'capacidad'] = np.nan
        
    if DB['plantilla_porcentaje'][i]>100:
        DB.loc[i,'plantilla_porcentaje'] = np.nan
        
    if DB['pago_impuestos'][i] == "No contesto":
        DB.loc[i,'pago_impuestos'] = np.nan
        
    if DB['pago_creditos'][i] == "No contesto":
        DB.loc[i,'pago_creditos'] = np.nan
        
    if DB['pago_renta'][i] == "No contesto":
        DB.loc[i,'pago_renta'] = np.nan
        
    if DB['pago_internet'][i] == "No contesto":
        DB.loc[i,'pago_internet'] = np.nan
        
    if DB['pago_agua'][i] == "No contesto":
        DB.loc[i,'pago_agua'] = np.nan
        
    if DB['pago_luz'][i] == "No contesto":
        DB.loc[i,'pago_luz'] = np.nan
        
    if DB['pago_gas'][i] == "No contesto":
        DB.loc[i,'pago_gas'] = np.nan
        
    if DB['pago_proporcion'][i]>120:
        DB.loc[i,'pago_proporcion'] = np.nan
        
    if DB['capacidad_pago_credito'][i] == "No contesto":
        DB.loc[i,'capacidad_pago_credito'] = np.nan
        
    if DB['credito_prorroga'][i] == "No contesto":
        DB.loc[i,'credito_prorroga'] = np.nan
    if DB['credito_prorroga'][i] == 998:
        DB.loc[i,'credito_prorroga'] = np.nan
    if DB['credito_prorroga'][i] == 999:
        DB.loc[i,'credito_prorroga'] = np.nan
        
    if DB['informado'][i] == "No contesto":
        DB.loc[i,'informado'] = np.nan
        
    if DB['lineamientos'][i] == "No contesto":
        DB.loc[i,'lineamientos'] = np.nan   
```

```python
#Sector vs Conocimiento_apoyo

import plotly.express as px
import pandas as pd

schools = ["Brown", "NYU", "Notre Dame", "Cornell", "Tufts", "Yale",
           "Dartmouth", "Chicago", "Columbia", "Duke", "Georgetown",
           "Princeton", "U.Penn", "Stanford", "MIT", "Harvard"]
n_schools = len(schools)

men_salary = [72, 67, 73, 80, 76, 79, 84, 78, 86, 93, 94, 90, 92, 96, 94, 112]
women_salary = [92, 94, 100, 107, 112, 114, 114, 118, 119, 124, 131, 137, 141, 151, 152, 165]

df = pd.DataFrame(dict(school=schools*2, salary=men_salary + women_salary,
                       gender=["Men"]*n_schools + ["Women"]*n_schools))

# Use column names of df for the different parameters x, y, color, ...
fig = px.scatter(df, x="salary", y="school", color="gender",
                 title="Gender Earnings Disparity",
                 labels={"salary":"Annual Salary (in thousands)"} # customize axis label
                )

fig.show()
```
