# Cálculo de la variable estrés

El cálculo de la variable se hará según la escala adaptada a cada una de las variables contenidas en la base de datos y que fueron seleccionadas como estresores económicos.

Ver documento de Word: Métrica Estresor Económico

```python
#Importar Librerías 
import pandas as pd 
import numpy as np
```

### Base de datos: IIEG_E_1

```python
#Lectura de la base de datos
data=pd.read_excel('data.xlsx', sheet_name='IIEG_E_1') 
#Acomodo de datos
data = data.replace({"Más de 52": 52, "De 26 a 52": 26, "No aplica": 100, "No contesto": 101, "No sé":102, "Más de un año":12})
#Variable a predecir 
estres = np.zeros(len(data))
```
